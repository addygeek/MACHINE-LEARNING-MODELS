```python
#importing the required library 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#importing the csv file into pandas dataframe
logistic = pd.read_csv("User_Data.csv")
```

```python
logistic.head()
```

```python
#lets spilit into X and Y
x = logistic.iloc[:,[2,3]].values
x
```

```python
y = logistic.iloc[:,-1].values
y
```

```python
#Import spliting the above x and y into traning and testing data
from sklearn.model_selection import train_test_split
```

```python
#Spliting data into training (x and y) and testing where 80% for training and 20% testing
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
x_old_train = x_train
```

```python
print('The total size of data is ',len(logistic))
print('X traning data and X testing datas are ',len(x_train),'and', len(x_test))
print('Y target data and Y testing datas are ',len(y_train),'and', len(y_test))
```

```python
#Lets import StandardScaler model from sklearn libray
from sklearn.preprocessing import StandardScaler
x_old_train
```

```python
#StandardScaler as Object
std_x = StandardScaler()
```

```python
#StandardScaler is used to scale the numbers in range so easy to interprete max-min techniques 
x_train = std_x.fit_transform(x_train)
```

```python
# X trained is now scaled downed in range
x_train
```

```python
#Similar doing in testing data
x_test = std_x.fit_transform(x_test)
```

```python
#Lets import the Logistic Regression module here to anlysis
from sklearn.linear_model import LogisticRegression
```

```python
#Storing the logistic regression object in varible "Log"
Log = LogisticRegression()
```

```python
#Finally Tranning the data
Log.fit(x_train,y_train)
```

```python
#Y prediction based on x are as below, so later do on other testing set as well after training
y_prediction = Log.predict(x_test)
y_prediction
```

```python
#Drwa Confusion Matrix
from sklearn.metrics import confusion_matrix
```

```python
#Lets fit then y predicted vs y test data into CM
cm = confusion_matrix(y_prediction, y_test)
print(cm)
```
