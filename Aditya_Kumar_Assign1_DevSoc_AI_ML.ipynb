```python
import numpy as np

# Created two 2x3 matrices
matrix_1 = np.arange(1, 7).reshape(2, 3)
matrix_2 = np.arange(7, 13).reshape(2, 3)

print("Matrix 1 \n", matrix_1)
print("Matrix 2 \n", matrix_2)

# Get the shape of the matrices
rows, cols = matrix_1.shape

# Initialize result matrices for addition and multiplication
added_matrix = np.zeros((rows, cols))
multiplied_matrix = np.zeros((rows, cols))

# Element-wise Addition
print("Performing Element-wise Addition:")
for i in range(rows):
    for j in range(cols):
        added_matrix[i][j] = matrix_1[i][j] + matrix_2[i][j]

print("Added Resultant Matrix: \n", added_matrix)

# Element-wise Multiplication
print("Performing Element-wise Multiplication:")
for i in range(rows):
    for j in range(cols):
        multiplied_matrix[i][j] = matrix_1[i][j] * matrix_2[i][j]

print("Element-wise Multiplied Matrix: \n", multiplied_matrix)
```

```python
#Question 4: Create a 4x4 identity matrix and multiply it with a 4x1 vector [ 1 , 2 , 3 , 4 ]
import numpy as np

# Create a 4x4 identity matrix
identity_matrix = np.eye(4)

# Create a 4x1 vector
vector = np.array([1, 2, 3, 4]).reshape(4, 1)

print("4x4 Identity Matrix:\n", identity_matrix)
print("4x1 Vector:\n", vector)

# Matrix multiplication
result_matrix = np.zeros((4, 1))

for i in range(4):
    for j in range(1):
        sum = 0
        for k in range(4):
            sum += identity_matrix[i][k] * vector[k][j]
        result_matrix[i][j] = sum

print("Result of Multiplication:\n", result_matrix)
```

```python
#Question 5: Create a 5x5 matrix of integers from 0 to 24. Extract the sub-matrix from the second to fourth rows and columns.
import numpy as np

# Create a 5x5 matrix with integers from 0 to 24
matrix = np.arange(25).reshape(5, 5)

print("Original 5x5 Matrix:\n", matrix)

# Extract the sub-matrix from the second to fourth rows and columns
sub_matrix = matrix[1:4, 1:4]

print("Extracted Sub-Matrix:\n", sub_matrix)
```

```python
#Question 7: Plot a sine wave using numpy and matplotlib for values ranging from 0 to 2π.
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate an array of values from 0 to 2π
x = np.linspace(0, 2 * np.pi, 1000)  # 1000 points between 0 and 2π

# Step 2: Compute the sine of these values
y = np.sin(x)

# Step 3: Lets Plot the sine wave
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Sine Wave')  # Plot the sine wave
plt.title('Sine Wave from 0 to 2pi')  # Title of the plot
plt.xlabel('x')  # X-axis label
plt.ylabel('sin(x)')  # Y-axis label
plt.grid(True)  # Add grid
plt.legend()  # Add legend
plt.show()  # Display the plot
```

```python
#Question 8: Create a scatter plot of random data points. The x-values should range from 0 to 50, and the y-values should be random integers between 10 and 100.
import matplotlib.pyplot as plt
import numpy as np

# Generate random x-values and y-values
np.random.seed(15)
x_values = np.random.randint(0, 51, 50)  # 50 random integers between 0 and 50
y_values = np.random.randint(10, 101, 50)  # 50 random integers between 10 and 100

# Created the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(x_values, y_values, c='blue', marker='o', edgecolor='k')

# Add titles and labels
plt.title('Scatter Plot of Random Data Points')
plt.xlabel('X Values')
plt.ylabel('Y Values')

# Display the plot
plt.grid(True)
plt.show()
```

```python
#Question 9: Plot the functions y = x^2 and y = x^3 on the same graph with a legend, different colors, and line styles.
import matplotlib.pyplot as plt
import numpy as np

# Generate x-values
x = np.linspace(-10, 10, 400)  # 400 points between -10 and 10

# Compute y-values for both functions
y1 = x**2
y2 = x**3

# Create the plot
plt.figure(figsize=(10, 6))

# Plot y = x^2
plt.plot(x, y1, label='$y = x^2$', color='blue', linestyle='-', linewidth=2)

# Plot y = x^3
plt.plot(x, y2, label='$y = x^3$', color='red', linestyle='--', linewidth=2)

# Add titles and labels
plt.title('Plot of $y = x^2$ and $y = x^3$ together')
plt.xlabel('x')
plt.ylabel('y')

# Add legend
plt.legend()

# Display grid
plt.grid(True)

# Show the plot
plt.show()
```

```python
#Question 10: Create a bar chart for the categorical data: Categories = ['A', 'B', 'C', 'D'],
#and Values = [4, 7, 1, 8].
import matplotlib.pyplot as plt

# Data for the bar chart
categories = ['A', 'B', 'C', 'D']
values = [4, 7, 1, 8]

# Create the bar chart
plt.figure(figsize=(8, 5))
bars = plt.bar(categories, values, color='orange', edgecolor='black')

# Add titles and labels
plt.title('Bar Chart of Categorical Data')
plt.xlabel('Categories')
plt.ylabel('Values')

# Add labels above bars
for bar in bars:
  height = bar.get_height()
  plt.text(bar.get_x() + bar.get_width() / 2, height, height, ha='center', va='bottom')

# Display the plot
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
#Question 1 has been done for you.

import pandas as pd
# Load Titanic dataset
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)
# Display the first five rows
df.head()
```

```python
#Question 2: Print a summary of the dataset, including the number of rows, columns, and data types.
print("SHAPE ", df.shape)
print("Number of rows:", df.shape[0])
print("Number of columns:", df.shape[1])
print("Data types of each column:\n", df.dtypes)
```

```python
#Question 3: Check for missing values in the dataset.
missing_values = df.isnull().sum()
print("Missing values in each column:\n", missing_values)
```

```python
#Question 4: Drop the 'Cabin' column and fill missing values in 'Age' with the median age.
# Drop the 'Cabin' column
df.drop(columns=['Cabin'], inplace=True)

# Fill missing values in 'Age' with the median age
df['Age'].fillna(df['Age'].median(), inplace=True)

# Verify changes
print("Missing values after cleaning:\n", df.isnull().sum())
```

```python
# prompt: Question 5: Count the number of passengers who survived and did not survive.

# Count the number of passengers who survived and did not survive
survived_count = df['Survived'].value_counts()
print("Survival Count:\n", survived_count)
```

```python
#Question 6: Calculate the average age of passengers who survived and those who did not.
average_age = df.groupby('Survived')['Age'].mean()
print("Average age of passengers who survived and did not survive:\n", average_age)
```

```python
#Question 7: Group by 'Pclass' and calculate the average fare paid by each class.
# Calculate the average fare for each passenger class
average_fare = df.groupby('Pclass')['Fare'].mean()
print("Average fare paid by each class:\n", average_fare)
```

```python
#Question 8: Determine the number of survivors in each passenger class.
survivors_by_class = df[df['Survived'] == 1].groupby('Pclass').size()
print("Number of survivors in each passenger class:\n", survivors_by_class)
```

#Data Visualization

```python
#Question 9: Plot a histogram of the 'Age' column to show the distribution of passengers' ages.
import matplotlib.pyplot as plt

# Plot histogram of passenger ages
plt.figure(figsize=(10, 6))
plt.hist(df['Age'], bins=30, color='red', edgecolor='black')
plt.title('Distribution of Passenger Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
#Question 10: Create a bar chart showing the number of passengers in each class.
# Plot a bar chart of the number of passengers in each class
plt.figure(figsize=(10, 6))
df['Pclass'].value_counts().sort_index().plot(kind='bar', color='lightgreen', edgecolor='black')
plt.title('Number of Passengers in Each Class')
plt.xlabel('Passenger Class')
plt.ylabel('Number of Passengers')
plt.xticks(ticks=[0, 1, 2], labels=['1st Class', '2nd Class', '3rd Class'], rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
#Question 11: Create a new column 'FamilySize' and find the average family size for each class.
# Create 'FamilySize' column
df['FamilySize'] = df['SibSp'] + df['Parch']

# Calculate the average family size for each class
average_family_size = df.groupby('Pclass')['FamilySize'].mean()
print("Average family size for passengers in each class:\n", average_family_size)
```

```python
# Question 12: Determine the correlation between 'Fare' and 'Age' and plot a scatter plot.
# Calculate the correlation between 'Fare' and 'Age'
correlation = df['Fare'].corr(df['Age'])
print("Correlation between 'Fare' and 'Age':", correlation)

# Plot scatter plot of 'Fare' vs. 'Age'
plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['Fare'], alpha=0.5, color='coral', edgecolor='k')
plt.title('Scatter Plot of Fare vs. Age')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.grid(True)
plt.show()
```

```python
#<Lets Plot the survival rate by passenger class.
# Calculate survival rates by passenger class
survival_rate_by_class = df.groupby('Pclass')['Survived'].mean()

# Plot survival rates by passenger class
plt.figure(figsize=(8, 5))
survival_rate_by_class.plot(kind='bar', color='salmon', edgecolor='black')
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.xticks(ticks=[0, 1, 2], labels=['1st Class', '2nd Class', '3rd Class'], rotation=0)
plt.ylim(0, 1)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Lets  Plot the number of passengers who survived and did not survive by gender.
# Count survivors and non-survivors by gender
survival_by_gender = df.groupby(['Sex', 'Survived']).size().unstack()

# Plot number of passengers by gender and survival status
plt.figure(figsize=(10, 6))
survival_by_gender.plot(kind='bar', stacked=True, color=['lightblue', 'lightcoral'], edgecolor='black')
plt.title('Number of Passengers Survived and Not Survived by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Passengers')
plt.xticks(ticks=[0, 1], labels=['Female', 'Male'], rotation=0)
plt.legend(['Did Not Survive', 'Survived'])
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
#Lets Plot a box plot of the 'Fare' distribution for each passenger class.
# Plot box plot of 'Fare' distribution by passenger class
plt.figure(figsize=(10, 6))
df.boxplot(column='Fare', by='Pclass', grid=False, patch_artist=True,
           boxprops=dict(facecolor='lightblue', color='black'),
           whiskerprops=dict(color='black'),
           medianprops=dict(color='red'))
plt.title('Fare Distribution by Passenger Class')
plt.suptitle('')  # Suppress the default title to avoid duplication
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.xticks(ticks=[1, 2, 3], labels=['1st Class', '2nd Class', '3rd Class'])
plt.grid(True)
plt.show()
```

```python
#Lets Plot a count plot of 'SibSp' (siblings/spouses aboard) and 'Parch' (parents/children aboard).
import seaborn as sns

# Plot count plot for 'SibSp'
plt.figure(figsize=(12, 6))
sns.countplot(x='SibSp', data=df, palette='viridis')
plt.title('Number of Siblings/Spouses Aboard')
plt.xlabel('Number of Siblings/Spouses')
plt.ylabel('Count')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Plot count plot for 'Parch'
plt.figure(figsize=(12, 6))
sns.countplot(x='Parch', data=df, palette='viridis')
plt.title('Number of Parents/Children Aboard')
plt.xlabel('Number of Parents/Children')
plt.ylabel('Count')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
#Lets Plot a pair plot to visualize relationships between 'Age', 'Fare', and 'Survived'.
# Select relevant columns for the pair plot
df_pairplot = df[['Age', 'Fare', 'Survived']]

# Plot pair plot
plt.figure(figsize=(10, 6))
sns.pairplot(df_pairplot, hue='Survived', palette='coolwarm', plot_kws={'alpha':0.6})
plt.title('Pair Plot of Age, Fare, and Survival')
plt.show()
```

```python
#Plot a heatmap of correlations between numerical features.
# Compute correlation matrix
correlation_matrix = df[['Age', 'Fare', 'SibSp', 'Parch']].corr()

# Plot heatmap of correlations
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, center=0)
plt.title('Heatmap of Feature Correlations')
plt.show()
```
